```python
import re
from nltk.corpus import stopwords
nltk.download('stopwords')

def clean_text(text):
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\@w+|\#','', text)
    text = re.sub(r'[^A-Za-z\s]', '', text)
    text = text.lower()
    return text

# Assuming 'text' column exists in your DataFrame for sentiment analysis
# Replace 'text' with the actual column name containing the text data if different
df['clean_text'] = df['text'].apply(clean_text)
```

```python
positive_text = " ".join(df[df['sentiment']=='positive']['clean_text'])
negative_text = " ".join(df[df['sentiment']=='negative']['clean_text'])

wc = WordCloud(width=800, height=400).generate(positive_text)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("Positive Sentiment Word Cloud")
plt.show()
```

```python
X = df['clean_text']
y = df['sentiment']

vectorizer = CountVectorizer()
X_vec = vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# Sentiment Analysis on Twitter Data in Google Colab

# Install required libraries
!pip install pandas matplotlib seaborn wordcloud nltk scikit-learn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import nltk
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import re

# Download NLTK stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords

# Load the dataset (make sure to upload your CSV file)
from google.colab import files
uploaded = files.upload()

# Replace 'your_file.csv' with the actual uploaded file name
df = pd.read_csv(list(uploaded.keys())[0])
df.head()

# Clean the text
def clean_text(text):
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\@\w+|\#','', text)
    text = re.sub(r'[^A-Za-z\s]', '', text)
    text = text.lower()
    return text

df['clean_text'] = df['tweet'].apply(clean_text)

# Visualize sentiment distribution
sns.countplot(x='sentiment', data=df)
plt.title("Sentiment Distribution")
plt.show()

# Word clouds for positive and negative sentiments
positive_text = " ".join(df[df['sentiment']=='positive']['clean_text'])
negative_text = " ".join(df[df['sentiment']=='negative']['clean_text'])

wc = WordCloud(width=800, height=400).generate(positive_text)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("Positive Sentiment Word Cloud")
plt.show()

wc = WordCloud(width=800, height=400).generate(negative_text)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("Negative Sentiment Word Cloud")
plt.show()

# Vectorize text and build model
X = df['clean_text']
y = df['sentiment']

vectorizer = CountVectorizer()
X_vec = vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
